```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
from scipy import stats
```

```python
import csv
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), stats.sem(a)
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 5))
propotion_of_bike_list=[0.3,0.4,0.5,0.6] 
bike_all_errors=[]
bike_all_errors_st=[]
for proportion in propotion_of_bike_list:
    s='simdata/ebikes_'+str(proportion)+'.csv'
    a= pd.read_csv(s)
    b=list((a.lost_demand)+(a.ebike_return_error)+(a.out_of_battery_trips))[15:]
    average_five_week  = []
    for i in range(7):
        average_five_week.append(sum(b[(5*i):(5*i+5)])/5)
    mean, error=mean_confidence_interval(average_five_week)
    bike_all_errors.append(mean)
    bike_all_errors_st.append(error)
#plt.plot(propotion_of_bike_list,bike_battery,'b',label = 'still happen')
plt.errorbar(propotion_of_bike_list,bike_all_errors,yerr=bike_all_errors_st,ecolor='y',capsize=3, color = 'b')    
plt.xlabel('ratio of ebikes')
plt.ylabel('avg total out-of-events per week')
plt.title('total out-of-events VS. ratio of ebikes')
plt.savefig('total_out-of-events_ebikes',dpi  = 300, bbox_inches = 'tight')
```
