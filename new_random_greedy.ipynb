```python
# random greedy
import Sim_random_pick_CRN_751
from datetime import timedelta as td
from datetime import datetime as time
import copy
import random
```

```python
start_time = time(2017,7,1,hour= 6)
end_time=start_time+ td(hours=10)
p = 0.3
N = int(751*p)
expected_battery_cost = eval(open(("expected_battery_cost.txt")).read())
initial_stations = eval(open(("steady_states_751/steady_ebikes_initial_0.5.txt")).read())
for station in initial_stations.keys():
    initial_stations[station]['edock'] = 0
subset = []
alpha = 0.2                 
for i in range(N):
    gc=Sim_random_pick_CRN_751.GlobalClock(start_time,end_time,initial_stations)
    gc.clockAdvance()
    # min_dif stores the minimum st_avg_soc - E(B)
    min_dif, minst = 101, -1
    # st_dif = st_avg_soc - E(B)
    st_dif = {}
    for key,st in gc.stations.items():
        sum = 0
        if len(st.ebike) == 0: continue
        for bike in st.ebike.values():
            sum += bike.SOC
        dif = sum / len(st.ebike) - expected_battery_cost[key]
        st_dif[key] = dif 
        #if dif < min_dif and key not in subset:
            #min_dif = dif
            #minst = key                       
    rcl = []
    sorted_list = sorted(st_dif.items(), key=lambda kv: kv[1])
    j = 0
    while (len(rcl) < 10):
        while sorted_list[j][0] in rcl or sorted_list[j][0] in subset:
            j += 1
        rcl.append(sorted_list[j][0])
    print(rcl)
    select = random.choice(rcl)
    select = random.choice(rcl)
    subset.append(select)
    initial_stations[select]['edock'] = initial_stations[select]['ecap']
    print(len(subset))
```
